```
import nltk

cfd = nltk.ConditionalFreqDist(
  (target, fileid[:4])
  for fileid in nltk.corpus.state_union.fileids()
  for w in nltk.corpus.state_union.words(fileid)
  for target in ['men', 'women', 'people']
  if w.lower().startswith(target))

cfd.plot()
```

```
from nltk.corpus import state_union
fileid = nltk.corpus.state_union.fileids()
pk = ['men', 'women', 'people']
for g in fileid:
  text = state_union.words(g)
  fdist = nltk.FreqDist(w.lower() for w in text)
  print('For fileid : '+g)
  for m in pk:
    print(m + ':',fdist[m],end=' ')
  print('\n')
```

```
from nltk.corpus import stopwords
most_freq_fd=nltk.FreqDist(brown.words(categories='news'))
#fd that includes stop words
print(most_freq_fd.most_common(50))
words=[word for word in most_freq_fd]
for word in words:
    if word in stopwords.words('english') or not word.isalpha():
        most_freq_fd.pop(word)
#fd that excludes stop words
print(most_freq_fd.most_common(50))
```
